```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic employee data
num_samples = 100

experience = np.random.randint(1, 31, num_samples)                 # Years of experience (1 to 30)
education = np.random.randint(1, 5, num_samples)                   # Education Level (1 to 4)
age = np.random.randint(22, 60, num_samples)                       # Age (22 to 60)
job_level = np.random.randint(1, 11, num_samples)                  # Job Level (1 to 10)
city_type = np.random.randint(1, 4, num_samples)                   # City Type (1 to 3)
skills_score = np.random.randint(40, 101, num_samples)             # Skills Score (40 to 100)
```

```python
# Polynomial salary function with interaction terms
salary = (3000 +
          500 * experience +
          2000 * education +
          1500 * job_level +
          1000 * city_type +
          30 * skills_score +
          10 * experience * job_level -
          5 * age ** 2 / 100 +
          np.random.normal(0, 5000, num_samples))  # Adding noise

# Create DataFrame
employee_df = pd.DataFrame({
    'Experience': experience,
    'Education': education,
    'Age': age,
    'Job_Level': job_level,
    'City_Type': city_type,
    'Skills_Score': skills_score,
    'Salary': np.round(salary, 2)
})

# Save dataset to CSV
file_path = 'employee_salary_dataset.csv'
employee_df.to_csv(file_path, index=False)

file_path
```

```python


# Load the generated dataset
employee_df = pd.read_csv('employee_salary_dataset.csv')
# Set plot style
sns.set_theme(style="whitegrid")

# Visualize the relationship between experience and salary
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Experience', y='Salary', data=employee_df, hue='Education', palette='viridis', alpha=0.8)
plt.title('Experience vs Salary (Colored by Education Level)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(title='Education Level')
plt.show()

# Pairplot for key features and salary
sns.pairplot(employee_df, vars=['Experience', 'Age', 'Job_Level', 'Skills_Score', 'Salary'], diag_kind='kde', corner=True)
plt.show()

# Heatmap to show correlation between variables
plt.figure(figsize=(8, 6))
sns.heatmap(employee_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```
